```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def poly(x):
    return x**3+6*x**2-x+17

a=-2
b=5
ans=0
sumeation = 0       
poly_int = quad(poly , -2 , 5)
print("using function " , poly_int[0])


point = np.random.uniform(-2 , 5 ,100000)

for n in point:
    sumeation = sumeation + (b-a) * poly(n)
    
ans = 1/100000 * (sumeation)
np.mean(ans)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

N =100000
a,b = (50,50)
x_min , x_max = {0 , .55}
randx = np.random.uniform(x_min , x_max , N)
y = stats.beta.pdf(randx , a ,b)
beta_at_55 = stats.beta(a , b).cdf(.55)
mcarlo_width = x_max-x_min
mcarlo_rect_area = mcarlo_width * y.sum()
mcarlo_rect_avg = mcarlo_rect_area/N
print(f"Real value : {beta_at_55} ")
print(f"Integral value :  {mcarlo_rect_avg} ")
```

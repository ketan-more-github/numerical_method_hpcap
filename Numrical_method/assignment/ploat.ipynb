```python
import matplotlib.pyplot as plt


guess = 0.0
cube = 27
increment = 0.1
elipsion = 0.1
cnt = 0
mygess = []
myelipsion = []


for i in range(1 , 35 ,1):
    val = guess**3 - cube
    myelipsion.append(val*-1)
    if abs(val) >= elipsion:
        guess = guess + increment
        mygess.append(guess)
        cnt=cnt+1
        print( cnt, ")" , guess)
         

if abs(guess**3 - cube) >= elipsion:
    print("faild on the cude root of" , cube)
else:
    print("\n")
    print(  cnt, ")" ,  guess , "is the cude root of" , cube)

  
  
print(mygess)
print(myelipsion)
        
        
plt.plot( mygess)
plt.plot(myelipsion )  
plt.xlabel('x - mygess') 
plt.ylabel('y - myelipsion') 
  
# giving a title to my graph 
plt.title('My gess vs elipsion graph') 
  
plt.show() 
```

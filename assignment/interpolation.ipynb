```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

m = [[1,1,1,1] , [1,2,4 ,8] , [1,3,9,27] , [ 1 , 3.4 , 3.4**2 , 3.4**3]]
yv=[1.1 , 2.1 ,5,7]

a = np.matmul(np.linalg.inv(m) , yv)

# print(a)

x = sp.symbols('x')
y = a[0] + a[1]*x + a[2]*x**2 + a[3]*x**3
x_val = np.arange(0,4,0.1)
# print(x_val)
y_val=[]

for i in x_val:
    y_val.append(y.subs(x,i))

# print(y_val)
plt.plot(x_val , y_val)
plt.scatter([1,2,3,3.4] , [1.1 , 2.1 , 5 , 7] , c='r')
plt.title(sp.N(y,5))
plt.grid()
plt.show()    
       
```
